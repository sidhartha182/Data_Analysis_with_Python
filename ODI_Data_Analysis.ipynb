## <center>Introduction to Cricket</center>

Cricket is a bat-and-ball game played between two teams of eleven players on a field at the centre of which is a 22-yard (20-metre) pitch with a wicket at each end, each comprising two bails balanced on three stumps. The batting side scores runs by striking the ball bowled at the wicket with the bat (and running between the wickets), while the bowling and fielding side tries to prevent this (by preventing the ball from leaving the field, and getting the ball to either wicket) and dismiss each batter (so they are "out"). Means of dismissal include being bowled, when the ball hits the stumps and dislodges the bails, and by the fielding side either catching the ball after it is hit by the bat and before it hits the ground, or hitting a wicket with the ball before a batter can cross the crease in front of the wicket. When ten batters have been dismissed, the innings ends and the teams swap roles. The game is adjudicated by two umpires, aided by a third umpire and match referee in international matches.

Forms of cricket range from Twenty20, with each team batting for a single innings of 20 overs, to Test matches played over five days. Traditionally cricketers play in all-white kit, but in limited overs cricket they wear club or team colours. In addition to the basic kit, some players wear protective gear to prevent injury caused by the ball, which is a hard, solid spheroid made of compressed leather with a slightly raised sewn seam enclosing a cork core layered with tightly wound string.

The earliest reference to cricket is in South East England in the mid-16th century. It spread globally with the expansion of the British Empire, with the first international matches in the second half of the 19th century. The game's governing body is the International Cricket Council (ICC), which has over 100 members, twelve of which are full members who play Test matches. The game's rules, the Laws of Cricket, are maintained by Marylebone Cricket Club (MCC) in London. The sport is followed primarily in the Indian subcontinent, Australasia, the United Kingdom, southern Africa and the West Indies.[1] Women's cricket, which is organised and played separately, has also achieved international standard. The most successful side playing international cricket is Australia, which has won seven One Day International trophies, including five World Cups, more than any other country and has been the top-rated Test side more than any other country.

# Number of matches Played by Each Country

```python
matches_played1=pd.DataFrame(odi_data.pivot_table(index=['Team 1'],aggfunc='size'))
matches_played2=pd.DataFrame(odi_data.pivot_table(index=['Team 2'],aggfunc='size'))
matches_played1=matches_played1.reset_index()
matches_played2=matches_played2.reset_index()
matches_played1=matches_played1.set_axis(['Team','Number_of_matches_as_Team1'],axis=1)
matches_played2=matches_played2.set_axis(['Team','Number_of_matches_as_Team2'],axis=1)
#matches_played=pd.merge(matches_played1,matches_played2,on='Team')
#matches_played['Total_Matches_Played']=matches_played.Number_of_matches_as_Team1+matches_played.Number_of_matches_as_Team2
#matches_played=matches_played[['Team','Total_Matches_Played']]
#matches_played=(matches_played.sort_values(by=['Total_Matches_Played'], ascending=False))
match_dict={}
count1=0
count2=0
for c in matches_played1.Team:
    if c not in match_dict:
        match_dict[c]=matches_played1.Number_of_matches_as_Team1[count1]
    count1=count1+1
for d in matches_played2.Team:
    if d not in match_dict:
        match_dict[d]=matches_played2.Number_of_matches_as_Team2[count2]
    if d in match_dict:
        match_dict[d]=match_dict[d]+matches_played2.Number_of_matches_as_Team2[count2]
    count2=count2+1
matches_played=pd.DataFrame(list(match_dict.items()))
matches_played=matches_played.set_axis(['Team','Number_of_matches'],axis=1)
```

```python
from IPython.core.display import HTML
display(HTML(matches_played.to_html()))
```

```python
fig=plt.figure(figsize=(20,20))
plt.pie(matches_played.Number_of_matches,labels=matches_played.Team);
matplotlib.rcParams['font.size']=50
plt.title('Number of Matches played by different Countries')
plt.show();
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(9,5)
matplotlib.rcParams['figure.facecolor']='#00000000'
plt.xticks(rotation=90)
sns.barplot(matches_played.Team,matches_played.Number_of_matches);
plt.title("Number of Matches Played(Country Wise)");
```

# Win Percentage(%) of Each Team

```python
win_by_countries1=pd.DataFrame(odi_data.pivot_table(index=['Winner'],aggfunc='size'))
win_by_countries1=win_by_countries1.reset_index()
win_by_countries1=win_by_countries1.set_axis(['Team','Number_of_matches_won'],axis=1)
win_by_countries1=win_by_countries1.drop([23,24])
win_by_countries1=pd.merge(win_by_countries1,matches_played,on='Team')
win_by_countries1['Win_Percentage']=(((win_by_countries1.Number_of_matches_won)/(win_by_countries1.Number_of_matches))*100)
```
